2) Vytvoř tabulku malé násobilky jako DataFrame (t.j. index i sloupce tvoří čísla 1-10, na průsečících je hodnota součinu).

```python
n = 13

columns = {}

for column in range(1, n + 1):
    columns[column] = [column * row for row in range(1, n + 1)]
    
multiplications = pd.DataFrame(columns, index = columns.keys())
multiplications
```

3) Napiš funkci multiplication_table(n), která vytvoří čtvercovou DataFrame s násobilkou s volitelným počtem řádků a sloupců n.(Stačí jen malinko adaptovat předchozí příklad).

```python
def multiply(n):
    columns = {}

    for column in range(1, n + 1):
        columns[column] = [column * row for row in range(1, n + 1)]
    return columns
```

```python
def multiplication_table(multiplications):
    return pd.DataFrame(multiplications, index = multiplications.keys())
```

```python
multiplication_table(multiply(14))
```

4) Použij tabulku pokémonů z hodiny. Pomocí tzv. indexerů zjisti: jaké hodnoty útoku a obrany má pokémon pikachu?

```python
def get_pokemons_values(
    table_name : str, 
    column : str, 
    pokemons : list, 
    attributes : list):
    
    pokemons_table = pd.read_csv(table_name)
    pokemons_table = pokemons_table.set_index(column).sort_index()
    results = pokemons_table.loc[pokemons, attributes]
    return results
```

```python
get_pokemons_values("pokemon.csv", "name", ["pikachu", "abra"], ["attack", "defense"])
```

5) Příště budeme pracovat s tabulkou zemí světa. Stáhni si tabulku jako csv z této adresy a:
    - podívej se, jaké obsahuje sloupce
    - zkus jí nastavit index podle jména a seřaď řádky podle indexu
    - vypiš řádky se zeměmi začínajícími na “Slo” a (subjektivně) je porovnej.

```python
def donwload_table(table_ulr, target_file_name):
    data = pd.read_csv(table_ulr)
    data.to_csv(target_file_name, index=False)
    return target_file_name
```

```python
def find_countries_beginning_with(string, target_file_name):
    countries_table = pd.read_csv(target_file_name)
    countries_ordered = countries_table.set_index("name").sort_index()
    return countries_ordered.loc[countries_ordered.index.str.startswith(string)]
```

```python
countries_results = find_countries_beginning_with(
    "Slo", 
    donwload_table(
        "https://raw.githubusercontent.com/janpipek/data-pro-pyladies/master/data/countries.csv",
        "countries.csv"
    )
)
countries_results
```
